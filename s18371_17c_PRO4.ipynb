```python
import numpy as np;
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

for i in range(1,17):
  print(i)
  #metoda Fisher'a z wyk≈Çadu
  #wczytanie x i y do macierzy
  file = np.loadtxt('/content/drive/MyDrive/Dane/dane{}.txt'.format(i))
  x=file[:,[0]]
  y=file[:,[1]]
  #utworzenie macierzy c z macierzy x orax macierzy 1
  c1=np.hstack([x, np.ones(x.shape)])
  #dla kwadratoweych podstawiamy x, x2 i macierz 1
  c3=np.hstack([x**2, x, np.ones(x.shape)])

  #macierz pseudoodwrotna od macierzy fi i y
  v1 = np.linalg.pinv(c1) @ y
  v3 = np.linalg.pinv(c3) @ y

  plt.plot(x, y, 'ro')
  plt.plot(x,v1[0]*x + v1[1])
  plt.plot(x,v3[0]*x**2 + v3[1]*x + v3[2])
  plt.show()
```
